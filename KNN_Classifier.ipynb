```python
# import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
# import dataset
ds= pd.read_csv("Social_Network_Ads.csv")
ds
```

```python
# finding duplicate
ds.duplicated().sum()
```

```python
# removing 
ds.drop_duplicates(inplace= True)
```

```python
# spliting dataset into independent and dependent var
x= ds[['Age', 'EstimatedSalary']]
y= ds[['Purchased']]
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
x= sc.fit_transform(x)
```

```python
# Model selection and fitting on Training test data

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
# Classification model fitting

from sklearn.neighbors import KNeighborsClassifier
cls =KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
cls.fit(x_train,y_train)
```

```python
# predict for myinput

cls.predict(sc.transform([[40,95000]]))
```

```python
# accuracy score

from sklearn.metrics import accuracy_score

y_pre =cls.predict(x_test)
accuracy= accuracy_score(y_test,y_pre)
print(f"Accuracy :{accuracy*100 :.2f}")
```

```python
# user input

age =float(input("Enter Age :"))
salary =float(input("Enter salary :"))

myinput= pd.DataFrame(sc.transform([[age,salary]]),columns =["Age","EstimatedSalary"])
result=cls.predict(myinput)

if result[0]==1:
    print("Prediction : User will purchased plan")
else:
    print("Prediction : User will not purchased plan")
```
